```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
import json

def encrypt_file(filename, qkd_key):
    # Open the JSON file and read the contents
    with open(filename, 'r') as file:
        data = json.load(file)
    
    # Convert the JSON data to a string
    plaintext = json.dumps(data)
    
    # Convert the QKD key to a byte string
    key = bytes.fromhex(qkd_key)
    
    # Generate a random initialization vector (IV) for AES
    iv = get_random_bytes(AES.block_size)
    
    # Encrypt the data using AES
    cipher = AES.new(key, AES.MODE_CBC, iv)
    ciphertext = iv + cipher.encrypt(pad(plaintext.encode('utf-8'), AES.block_size))
    
    # Write the encrypted data to a new file
    with open(filename + '.enc', 'wb') as file:
        file.write(ciphertext)

qkd_key = generate_key()
encrypt_file('data.json', qkd_key)


def decrypt_file(filename, qkd_key):
    # Convert the QKD key to a byte string
    key = bytes.fromhex(qkd_key)
    
    # Open the encrypted file and read the contents
    with open(filename, 'rb') as file:
        ciphertext = file.read()
    
    # Extract the initialization vector (IV) from the ciphertext
    iv = ciphertext[:AES.block_size]
    
    # Decrypt the data using AES
    cipher = AES.new(key, AES.MODE_CBC, iv)
    plaintext = cipher.decrypt(ciphertext[AES.block_size:])
    
    # Remove padding
    plaintext = unpad(plaintext, AES.block_size)
    
    # Convert the decrypted data to a JSON string
    data = json.loads(plaintext.decode('utf-8'))
    
    # Print the decrypted data
    print(data)
    
    # Write the decrypted data to a new file
    with open('dec_data.json', 'w') as file:
        file.write(json.dumps(data))

decrypt_file('data.json.enc', qkd_key)
```
